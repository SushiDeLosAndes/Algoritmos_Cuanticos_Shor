```python
""" Importar paquetes y clases de Qiskit"""
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit import execute, IBMQ
from qiskit import BasicAer, Aer
import sys

""" Importar funciones de Python """
import math
import array
import fractions
import numpy as np
```

```python
""" Función para verificar si N es de la forma q^p"""
def check_if_power(N):
    """ Determinar si N es una potencia perfecta en tiempo polinomial O(n^3), donde n = ⌈log_2 N⌉ """
    b=2
    while (2**b) <= N:
        a = 1
        c = N
        while (c-a) >= 2:
            m = int( (a+c)/2 )

            if (m**b) < (N+1):
                p = int( (m**b) )
            else:
                p = int(N+1)

            if int(p) == int(N):
                print('N es {0}^{1}'.format(int(m),int(b)) )
                return True

            if p<N:
                a = int(m)
            else:
                c = int(m)
        b=b+1

    return False
```

```python
""" Función para obtener el parámetro "a" ( 1<a<N ), tal que "a" y N son coprimos. Se inicializa por medio del "a" más pequeño
    El usuario tiene la elección de elegir el valor de "a", en vez de realizar el procedimiento aleatoriamente."""
def get_value_a(N):

    """ok define si el usuario desea usar el valor de "a" sugerido (si ok!='0') o no (if ok=='0')"""
    ok='0'

    """ Inicializar con a=2 """
    a=2

    """ Obtener el "a" más pequeño tal que "a" y N son coprimos """
    while math.gcd(a,N)!=1:
        a=a+1

    """ Guardar este valor como el "a" más pequeño """
    smallest_a = a

    """ Se pregunta al usuario si el "a" encontrado es bueno. Si no, se incrementa el valor de "a" y se pregunta nuevamente """  
    ok = input('El número {0} es bueno para a? Presione 0 si no, presione cualquier otro número de lo contrario: '.format(a))
    if ok=='0':
        if(N==3):
            print('El número {0} es el único número que se puede usar.'.format(a,a))
            return a
        a=a+1

    """ Se genera un ciclo para encontrar todas las posibilidades para "a", sin contar el "a" más pequeño, hasta que el usuario confirme un "a" como bueno """
    while ok=='0':
        
        """ Obtener "a" coprimo con respecto a N """
        while math.gcd(a,N)!=1:
            a=a+1
    
        """ Preguntar si este "a" es bueno """
        ok = input('El número {0} es bueno para a? Presione 0 si no, presione cualquier otro número de lo contrario: '.format(a))

        """ Si el usuario confirma que es bueno, entonces se sale del ciclo, ya que "a" ha sido encontrado """
        if ok!='0':
            break
        
        """ Si el usuario confirma que no es bueno, se incrementa "a" y se verifica que se han tomado todas las posibilidades.  """
        a=a+1

        """ Si todas las posibilidades para "a" han sido rechazadas, se toma "a" como el "a" más pequeño y se sale del ciclo """
        if a>(N-1):
            print('Se han rechazado todos los valores de a\n, seleccionando el más pequeño')
            a=smallest_a
            break

    """ Se imprime el valor de "a" """
    print('Se usará {0} como el valor de a\n'.format(a))

    return a
```

```python
""" Función para aplicar el algoritmo de las fracciones continuas para hallar el periodo r y el mínimo común divisor para encontrar los factores deseados """
def get_factors(x_value,t_upper,N,a):

    if x_value<=0:
        print('x_value es <= 0, no se puede aplicar el algoritmo de las fracciones continuas\n')
        return False

    print('Aplicando el algoritmo de las fracciones continuas\n')

    """ Calcular T y x/T """
    T = pow(2,t_upper)

    x_over_T = x_value/T

    """ Se realiza un ciclo en el que cada iteración corresponde a añadir un término más en el cálculo de la fracción continua (FC) de x/T """

    """ Inicializar los primeros valores correspondientes a las reglas de FC """
    i=0
    b = array.array('i')
    t = array.array('f')

    b.append(math.floor(x_over_T))
    t.append(x_over_T - b[i])

    while i>=0:

        """Desde la segunda iteración en adelante, se calculan los nuevos términos de FC a partir
        de los términos previos"""

        if i>0:
            b.append( math.floor( 1 / (t[i-1]) ) ) 
            t.append( ( 1 / (t[i-1]) ) - b[i] )

        """ Se calcula FC usando los términos conocidos """

        aux = 0
        j=i
        while j>0:    
            aux = 1 / ( b[j] + aux )      
            j = j-1
        
        aux = aux + b[0]

        """Obtener el denominador de los valores obtenidos"""
        frac = fractions.Fraction(aux).limit_denominator()
        den=frac.denominator

        print('Aproximación número {0} de la fracción continua:'.format(i+1))
        print("Numerador:{0} \t\t Denominador: {1}\n".format(frac.numerator,frac.denominator))

        """ Se incrementa i para la próxima iteración """
        i=i+1

        if (den%2) == 1:
            if i>=15:
                print('Se han realizado muchos intentos. Reiniciando el algoritmo de las fracciones continuas')
                return False
            print('Denominador impar. No es válido para los procesos subsecuentes. Se realizará una nueva iteración del algoritmo de la fracción continua\n')
            continue
    
        """ Si el denominador es par, se intentan obtener los factores de N """

        """ Se obtiene el factor exponencial a^(r/2) """

        exponential = 0

        if den<1000:
            exponential=pow(a , (den/2))
        
        """ Verificar si el valor es muy grande o no """
        if math.isinf(exponential)==1 or exponential>1000000000:
            print('Denominador de la fracción continua es muy grande.\n')
            aux_out = input('Ingrese el número 1 si se desea continuar la búsqueda. De lo contrario, ingrese otro número: ')
            if aux_out != '1':
                return False
            else:
                continue

        """Si el valor no es demasiado grande, se obtienen los valores y se llega al mínimo común divisor"""

        putting_plus = int(exponential + 1)

        putting_minus = int(exponential - 1)
    
        one_factor = math.gcd(putting_plus,N)
        other_factor = math.gcd(putting_minus,N)
    
        """ Verificar si los factores encontrados son triviales o son correctos """

        if one_factor==1 or one_factor==N or other_factor==1 or other_factor==N:
            print('Se encontraron factores triviales.\n')
            """ Verificar si el número ya había sido encontrado usando el intento i-1-ésimo """
            if t[i-1]==0:
                print('El algoritmo de las fracciones continuas ha encontrado x_final/(2^(2n)). Saliendo del algoritmo.\n')
                return False
            if i<15:
                aux_out = input('Ingrese el número 1 si se desea continuar la búsqueda. De lo contrario, ingrese otro número: ')
                if aux_out != '1':
                    return False       
            else:
                """ Se reinicia tras varios intentos y si los números son muy grandes.""" 
                print('Reiniciando el algoritmo, ya que se han realizado varios intentos.\n')
                return False         
        else:
            print('Los factores de {0} son {1} y {2}\n'.format(N,one_factor,other_factor))
            print('Se han hallado los factores deseados!\n')
            return True
```

```python
"Función correspondiente al mínimo común divisor."
def egcd(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        g, y, x = egcd(b % a, a)
        return (g, x - (b // a) * y, y)

"Función correspondiente al inverso modular."
def modinv(a, m):
    g, x, y = egcd(a, m)
    if g != 1:
        raise Exception('No existe el inverso modular.')
    else:
        return x % m
```

```python
""" Function para crear QFT """
def create_QFT(circuit,up_reg,n,with_swaps):
    i=n-1
    """ Se aplican compuertas H (Hadamard) and Cfases """
    """ Las Cfases con |ángulo| < "umbral" no serán creadas, ya que no aportan en el desarrollo de QFT. El umbral se determinará como 0,
    de manera que todas las Cfases son creadas, pero es modificable según desee el usuario.  """
    while i>=0:
        circuit.h(up_reg[i])        
        j=i-1  
        while j>=0:
            if (np.pi)/(pow(2,(i-j))) > 0:
                circuit.cu1( (np.pi)/(pow(2,(i-j))) , up_reg[i] , up_reg[j] )
                j=j-1   
        i=i-1  

    """ Si se específica, se aplican swaps (intercambios) al final de la función """
    if with_swaps==1:
        i=0
        while i < ((n-1)/2):
            circuit.swap(up_reg[i], up_reg[n-1-i])
            i=i+1
```

```python
""" Función para crear QFT inversa """
def create_inverse_QFT(circuit,up_reg,n,with_swaps):
    """ Si se específica, se aplican swaps (intercambios) al comienzo de la función"""
    if with_swaps==1:
        i=0
        while i < ((n-1)/2):
            circuit.swap(up_reg[i], up_reg[n-1-i])
            i=i+1
    
    """ Se aplican compuertas H (Hadamard) and Cfases """
    """ Las Cfases con |ángulo| < "umbral" no serán creadas, ya que no aportan en el desarrollo de QFT. El umbral se determinará como 0,
    de manera que todas las Cfases son creadas, pero es modificable según desee el usuario.  """
    
    i=0
    while i<n:
        circuit.h(up_reg[i])
        if i != n-1:
            j=i+1
            y=i
            while y>=0:
                 if (np.pi)/(pow(2,(j-y))) > 0:
                    circuit.cu1( - (np.pi)/(pow(2,(j-y))) , up_reg[j] , up_reg[y] )
                    y=y-1   
        i=i+1
```

```python
"""Función que devuelve el arreglo de ángulos a usar en la adición correspondiente en el espacio de Fourier"""
def getAngles(a,N):
    s=bin(int(a))[2:].zfill(N) 
    angles=np.zeros([N])
    for i in range(0, N):
        for j in range(i,N):
            if s[j]=='1':
                angles[N-i-1]+=math.pow(2, -(j-i))
        angles[N-i-1]*=np.pi
    return angles
```

```python
"""Creación de una compuerta de fase doblemente controlada"""
def ccphase(circuit,angle,ctl1,ctl2,tgt):
    circuit.cu1(angle/2,ctl1,tgt)
    circuit.cx(ctl2,ctl1)
    circuit.cu1(-angle/2,ctl1,tgt)
    circuit.cx(ctl2,ctl1)
    circuit.cu1(angle/2,ctl2,tgt)
    
"""Creación de un circuito que realice la adición con respecto a "a" en el espacio de Fourier"""
def phiADD(circuit,q,a,N,inv):
    angle=getAngles(a,N)
    for i in range(0,N):
        if inv==0:
            circuit.u1(angle[i],q[i])
        else:
            circuit.u1(-angle[i],q[i])

"""Versión controlada del circuito phiADD"""
def cphiADD(circuit,q,ctl,a,n,inv):
    angle=getAngles(a,n)
    for i in range(0,n):
        if inv==0:
            circuit.cu1(angle[i],ctl,q[i])
        else:
            circuit.cu1(-angle[i],ctl,q[i])

"""Versión doblemente controlada del circuito phiADD"""    
def ccphiADD(circuit,q,ctl1,ctl2,a,n,inv):
    angle=getAngles(a,n)
    for i in range(0,n):
        if inv==0:
            ccphase(circuit,angle[i],ctl1,ctl2,q[i])
        else:
            ccphase(circuit,-angle[i],ctl1,ctl2,q[i])
        
"""Circuito que implementa la adición modular doblemente controlada con respecto a "a" """
def ccphiADDmodN(circuit, q, ctl1, ctl2, aux, a, N, n):
    ccphiADD(circuit, q, ctl1, ctl2, a, n, 0)
    phiADD(circuit, q, N, n, 1)
    create_inverse_QFT(circuit, q, n, 0)
    circuit.cx(q[n-1],aux)
    create_QFT(circuit,q,n,0)
    cphiADD(circuit, q, aux, N, n, 0)
    
    ccphiADD(circuit, q, ctl1, ctl2, a, n, 1)
    create_inverse_QFT(circuit, q, n, 0)
    circuit.x(q[n-1])
    circuit.cx(q[n-1], aux)
    circuit.x(q[n-1])
    create_QFT(circuit,q,n,0)
    ccphiADD(circuit, q, ctl1, ctl2, a, n, 0)

"""Circuito que implementa la adición modular inversa doblemente controlada con respecto a "a" """
def ccphiADDmodN_inv(circuit, q, ctl1, ctl2, aux, a, N, n):
    ccphiADD(circuit, q, ctl1, ctl2, a, n, 1)
    create_inverse_QFT(circuit, q, n, 0)
    circuit.x(q[n-1])
    circuit.cx(q[n-1],aux)
    circuit.x(q[n-1])
    create_QFT(circuit, q, n, 0)
    ccphiADD(circuit, q, ctl1, ctl2, a, n, 0)
    cphiADD(circuit, q, aux, N, n, 1)
    create_inverse_QFT(circuit, q, n, 0)
    circuit.cx(q[n-1], aux)
    create_QFT(circuit, q, n, 0)
    phiADD(circuit, q, N, n, 0)
    ccphiADD(circuit, q, ctl1, ctl2, a, n, 1)
```

```python
"""Circuito que implementa la multiplicación modular controlada con respecto a "a" """
def cMULTmodN(circuit, ctl, q, aux, a, N, n):
    create_QFT(circuit,aux,n+1,0)
    for i in range(0, n):
        ccphiADDmodN(circuit, aux, q[i], ctl, aux[n+1], (2**i)*a % N, N, n+1)
    create_inverse_QFT(circuit, aux, n+1, 0)

    for i in range(0, n):
        circuit.cswap(ctl,q[i],aux[i])

    a_inv = modinv(a, N)
    create_QFT(circuit, aux, n+1, 0)
    i = n-1
    while i >= 0:
        ccphiADDmodN_inv(circuit, aux, q[i], ctl, aux[n+1], math.pow(2,i)*a_inv % N, N, n+1)
        i -= 1
    create_inverse_QFT(circuit, aux, n+1, 0)
```

```python

""" Programa principal """
if __name__ == '__main__':

    """ Se pregunta por el número N a analizar """   

    N = int(input('Inserte el número entero positivo N: '))

    print('El número ingresado es: {0}\n'.format(N))
    
    """ Verificar si N==1 o N==0"""

    if N==1 or N==0: 
       print('Ingrese un N distinto a 0 o 1')
       exit()
    
    """ Verificar si N es par """

    if (N%2)==0:
        print('N es par. La factorización es trivial.')
        exit()
    
    """ Verificar si N es de la forma N=p^q, p>1, q>=2 """

    """ Se intentan todos los números posibles para p: desde 2 hasta sqrt(N) """
    if check_if_power(N)==True:
       exit()

    print('La factorización no es trivial. Aplicando el circuito cuántico...\n')

    """ To login to IBM Q experience the following functions should be called """
    
    "IBMQ.delete_accounts()"
    "IBMQ.save_account('insert token here')"
    "IBMQ.load_accounts()"
    

    """ Se obtiene un entero positivo "a" que es coprimo con N """
    a = get_value_a(N)

    """ Obtener el valor de "n" para aplicar el algoritmo de Shor y determinar el número de cubits a utilizar """
    n = math.ceil(math.log(N,2))
    
    print('Número total de cubits a utilizar: {0}\n'.format(4*n+2))

    """ Creación de registros clásicos y cuánticos """

    """Registro cuántico auxiliar, usado en la multiplicación y la adición"""
    aux = QuantumRegister(n+2)
    """Registro cuántico donde se realiza QFT"""
    up_reg = QuantumRegister(2*n)
    """Registro cuántico donde se realizan las multiplicaciones"""
    down_reg = QuantumRegister(n)
    """Registro clásico donde los valores medidos en QFT son guardados"""
    up_classic = ClassicalRegister(2*n)

    """ Creación del circuito cuántico """
    circuit = QuantumCircuit(down_reg , up_reg , aux, up_classic)

    """ Se inicializa el registro clásico a 1 y se crea una superposición inicial en los registros cuánticos """
    circuit.h(up_reg)
    circuit.x(down_reg[0])

    """ Aplicando compuertas de multipliciación para crear la exponenciación """
    for i in range(0, 2*n):
        cMULTmodN(circuit, up_reg[i], down_reg, aux, int(pow(a, pow(2, i))), N, n)

    """ Aplicando QFT inversa"""
    create_inverse_QFT(circuit, up_reg, 2*n ,1)

    """ Midiendo los cubits superiores, obtieniendo el valor de x"""
    circuit.measure(up_reg,up_classic)

    """ Seleccionar la cantidad de veces que se ejecutará el circuito"""
    number_shots=int(input('Número de veces que se ejecutará el circuito: '))
    if number_shots < 1:
        print('Ejecute el circuito al menos una vez...')
        exit()

    if number_shots > 1:
        print('\n En el caso en el que el circuito tome mucho tiempo en completarse, considere en ejecutar el circuito un menor número de veces\n')

    """ Información que se imprime al usuario """
    print('Ejecución número {0} del circuito para N={1} y a={2}\n'.format(number_shots,N,a))

    """ Simulación del circuito cuántico creado """
    simulation = execute(circuit, backend=Aer.get_backend('aer_simulator'),shots=number_shots)

    """ to run on IBM, use backend=IBMQ.get_backend('ibmq_qasm_simulator') in execute() function """
    """ to run locally, use backend=BasicAer.get_backend('qasm_simulator') in execute() function """

    """ Obtener los resultados de la ejecución """
    sim_result=simulation.result()
    counts_result = sim_result.get_counts(circuit)

    """ Imprimir información de los resultados simulados al usuario """
    print('Imprimiendo los resultados y el número de intentos necesarios (con respecto a los {} casos):\n'.format(number_shots))
    i=0
    while i < len(counts_result):
        print('El resultado binario \"{0}\" ocurrió {1} vez/veces con respecto a los {2} casos'.format(list(sim_result.get_counts().keys())[i],list(sim_result.get_counts().values())[i],number_shots))
        i=i+1

    """ Impresión en blanco para dejar espacio entre resultados """
    print(' ')

    """ Inicialización de la probabilidad de éxito """
    prob_success=0

    """ Por cada resultado simulado, se entrega información correspondiente al usuario y se trata de clacular los factores de N"""
    i=0
    while i < len(counts_result):

        """ Obtieniendo el x_value de los cubits del estado final """
        output_desired = list(sim_result.get_counts().keys())[i]
        x_value = int(output_desired, 2)
        prob_this_result = 100 * ( int( list(sim_result.get_counts().values())[i] ) ) / (number_shots)

        print("------> Analizando el resultado número {0}. Este resultado se dio un {1:.4f} % de veces con respecto a todos los casos intentados\n".format(output_desired,prob_this_result))

        """ Imprimiendo x_final para el usuario """
        print('En forma decimal, el valor de x_final para este resultado es: {0}\n'.format(x_value))

        """ Obtieniendo los factores correspondientes a x_final"""   
        success=get_factors(int(x_value),int(2*n),int(N),int(a))
        
        if success==True:
            prob_success = prob_success + prob_this_result

        i=i+1

    print("\n Por medio de a={0}, se encontraron los factores of N={1} en el {2:.4f} % de los casos\n".format(a,N,prob_success))
```
